```python
from PIL import Image
import numpy as np
import json
from tqdm import tqdm
# Setup detectron2 logger
import detectron2
from detectron2.utils.logger import setup_logger
setup_logger()

# import some common libraries
import os
import numpy as np
import cv2
import random
from google.colab.patches import cv2_imshow

# import some common detectron2 utilities
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.engine import DefaultTrainer
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
```

## 이미지에서 사람을 찾기 (coco data model 사용)

```python
# coco data 기반 사람 찾기
coco_cfg = get_cfg()
coco_cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml"))
coco_cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5  # set threshold for this model
coco_cfg.MODEL.WEIGHTS = "https://dl.fbaipublicfiles.com/detectron2/COCO-Detection/faster_rcnn_R_101_FPN_3x/137851257/model_final_f6e8b1.pkl"
# coco_cfg.MODEL.WEIGHTS = "/content/drive/MyDrive/img00/output/model_final.pth"

# Create predictor
coco_predictor = DefaultPredictor(coco_cfg)
```

```python
!wget https://img-lcwaikiki.mncdn.com/mnresize/1024/-//productimages/20201/1/3945185/l_20201-0sg016z8-cs8_a.jpg -O /content/example.jpg
```

```python
from detectron2.utils.visualizer import ColorMode
im = cv2.imread("/content/example.jpg")
coco_outputs = coco_predictor(im)
print(coco_outputs)
# We can use `Visualizer` to draw the predictions on the image.
v = Visualizer(im[:, :, ::-1], MetadataCatalog.get(coco_cfg.DATASETS.TRAIN[0]), scale=1.2)
v = v.draw_instance_predictions(coco_outputs["instances"].to("cpu"))
cv2_imshow(v.get_image()[:, :, ::-1])
```

## 옷을 찾기 위한 deepfasion 모델

```python
# 옷 분류를 위한 리스트
lst_name = ['short_sleeved_shirt', 'long_sleeved_shirt', 'short_sleeved_outwear', 'long_sleeved_outwear',
            'vest', 'sling', 'shorts', 'trousers', 'skirt', 'short_sleeved_dress',
            'long_sleeved_dress', 'vest_dress', 'sling_dress']
```

```python
deepfasion_cfg = get_cfg()

deepfasion_cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml"))
deepfasion_cfg.DATASETS.TRAIN = ("deepfashion_val",)
deepfasion_cfg.DATASETS.TEST = ()
deepfasion_cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml")  # Let training initialize from model zoo

deepfasion_cfg.SOLVER.IMS_PER_BATCH = 4
deepfasion_cfg.SOLVER.BASE_LR = 0.001
deepfasion_cfg.SOLVER.WARMUP_ITERS = 1000
deepfasion_cfg.SOLVER.MAX_ITER = 1500
deepfasion_cfg.SOLVER.STEPS = (1000, 1500)
deepfasion_cfg.SOLVER.GAMMA = 0.05
deepfasion_cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 64
deepfasion_cfg.MODEL.ROI_HEADS.NUM_CLASSES = 13

deepfasion_cfg.TEST.EVAL_PERIOD = 500
```

```python
deepfasion_cfg.OUTPUT_DIR = r"/content/drive/MyDrive/playdata_project/data/model"
```

```python
def deep_fashion_dataset():
  return {k: v for k, v in enumerate(lst_name)}

from detectron2.data import DatasetCatalog
DatasetCatalog.register("deep_fashion2", deep_fashion_dataset)
```

```python
deepfasion_cfg.MODEL.WEIGHTS = os.path.join(deepfasion_cfg.OUTPUT_DIR, "model_final.pth")
deepfasion_cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.58   # set the testing threshold for this model
deepfasion_cfg.DATASETS.TEST = ("deep_fashion2")
deepfasion_predictor = DefaultPredictor(deepfasion_cfg)
```

```python
from detectron2.utils.visualizer import ColorMode
deepfasion_outputs = deepfasion_predictor(im)
# We can use `Visualizer` to draw the predictions on the image.
v = Visualizer(im[:, :, ::-1], MetadataCatalog.get("deep_fashion2"), scale=1.2)
v = v.draw_instance_predictions(deepfasion_outputs["instances"].to("cpu"))
cv2_imshow(v.get_image()[:, :, ::-1])
```

## 분류기 병합

```python
!unzip /content/drive/MyDrive/playdata_project/data/image/completed_original_insta_image.zip -d /content/drive/MyDrive/playdata_project/data/image/completed_original_insta_image
```

```python
import os
img_dir = '/content/drive/MyDrive/playdata_project/data/image/completed_original_insta_image'
for bn in tqdm(os.listdir(img_dir)) : # base_name
  fn = f'{img_dir}/{bn}'
  bn = os.path.splitext(bn)[0]

  coco_img = cv2.imread(fn)
  coco_outputs = coco_predictor(coco_img)
  coco_cls = coco_outputs['instances'].to('cpu').pred_classes.numpy()
  coco_boxes = coco_outputs['instances'].to('cpu').pred_boxes.tensor.numpy().astype(int)

  for i in range(coco_boxes.shape[0]) :
    b_cls = coco_cls[i]

    if b_cls != 0 :
      continue # if img is not persion, pass

    x1, y1, x2, y2 = coco_boxes[i]
    coco_cropped_img = coco_img[y1 : y2, x1 : x2]

    deepfasion_outputs = deepfasion_predictor(coco_cropped_img)
    deepfasion_cls = deepfasion_outputs['instances'].to('cpu').pred_classes.numpy()
    deepfasion_boxes = deepfasion_outputs['instances'].to('cpu').pred_boxes.tensor.numpy().astype(int)

    for j in range(deepfasion_boxes.shape[0]) :
      x1, y1, x2, y2 = deepfasion_boxes[j]
      cropped_img = coco_cropped_img[y1 : y2, x1 : x2]
      cv2.imwrite(f'//content/drive/MyDrive/playdata_project/data/image/completed_cropped_insta_image/{bn}_p{i}_b{j}_c{deepfasion_cls[j]}.jpg', cropped_img)

      





    # cv2.imwrite(f'/content/drive/MyDrive/img00/cropped_img3/{bn}_b{i}_c{cls[i]}.jpg', cropped_img)


```
