# 무신사용 pickle and annoy file

```python
# 무신사용
mu_dir = '../data/image/mall_image'
# f = 4096

f = 22
t = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed

mu_index = {k:v for k, v in enumerate(os.listdir(mu_dir))}

with open('../data/pickle_annoy/mu_index.pickle', 'wb') as f:
    pickle.dump(mu_index, f, pickle.HIGHEST_PROTOCOL)

ls = []
for no, fn in tqdm(mu_index.items()):
    try:
        img = cv2.imread('{}/{}'.format(mu_dir,fn))
        v = get_feature_vector1(img)[0]
        t.add_item(no, v)

    except:
        ls.append(fn)
        continue

t.build(10) # 10 trees
t.save('../data/pickle_annoy/mu_search.ann') # musinsa 검색용
```
